```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("C:\\Users\\saini\\Desktop\\depositphotos_2140353-stock-illustration-chessboard-black-and-white.jpg")
plt.imshow(img)
```

```python
gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap= "gray")
```

```python
gray = np.float32(gray)
```

```python
cd =cv2.cornerHarris(gray,2 ,3, 0.5)
dilate= cv2.dilate(cd,None)
plt.imshow(dilate)
```

```python
thresh = 1*dilate.max()
for i in range( 0, dilate.shape[0]):
    for j in range(0, dilate.shape[1]):
        if(dilate[i,j]>thresh):
            cv2.circle(img, (i,j),2, (214,255,0),1)
plt.imshow(img)
```
